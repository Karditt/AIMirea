**–†–µ—à–µ–Ω–∏–µ:**

```python
import matplotlib.pyplot as plt 
import seaborn as sns; sns.set() 
import numpy as np 
import pandas as pd
from sklearn.cluster import KMeans

X = np.array([[5,3],
              [10,15],
              [15,12],
              [24,10],
              [30,45],
              [85,70],
              [71,80],
              [60,78],
              [55,52],
              [80,91],]
            )
df = pd.DataFrame()
df['data'] = list(X)
kmeans = KMeans(n_clusters=3, random_state = 0)
clusters = kmeans.fit_predict(X)
df['cluster'] = clusters
colors = ['red', 'blue', 'green', 'black', 'yellow']
for row in df.to_dict('records'):
    plt.scatter(*row['data'], color=colors[row['cluster']])
plt.show()
```

```python
from sklearn.neighbors import KNeighborsClassifier
from matplotlib import pyplot as plt
import seaborn as sns
from sklearn.cluster import KMeans


iris = sns.load_dataset("iris")
X = np.array([[row['sepal_width'] ,row['sepal_length']] for row in iris.to_dict('records')])
X2 = np.array([[row['petal_width'] ,row['petal_length']] for row in iris.to_dict('records')])
colors = ['red', 'blue', 'green', 'black', 'yellow']
df = pd.DataFrame()
df['data'] = list(X)
kmeans = KMeans(n_clusters=4, random_state = 0)
clusters = kmeans.fit_predict(X)
df['cluster'] = clusters

df2 = pd.DataFrame()
df2['data'] = list(X2)
kmeans = KMeans(n_clusters=3, random_state = 0)
clusters = kmeans.fit_predict(X2)
df2['cluster'] = clusters

for row in df.to_dict('records'):
    plt.scatter(*row['data'], color=colors[row['cluster']])
plt.show()
for row in df2.to_dict('records'):
    plt.scatter(*row['data'], color=colors[row['cluster']])
plt.show()
```

**–†–µ—à–µ–Ω–∏–µ:**

```python
from scipy.cluster.hierarchy import dendrogram, linkage
from matplotlib import pyplot as plt
import seaborn as sns
from sklearn.cluster import AgglomerativeClustering
import pandas as pd

iris = sns.load_dataset("iris")
X = np.array([[row['sepal_width'] ,row['sepal_length']] for row in iris.to_dict('records')])
X2 = np.array([[row['petal_width'] ,row['petal_length']] for row in iris.to_dict('records')])

cluster = AgglomerativeClustering(n_clusters = 6, affinity = 'euclidean', linkage = 'ward')
cluster.fit_predict(X)
plt.scatter(X[:,0],X[:,1], c = cluster.labels_, cmap = 'rainbow')
plt.show()

linkage_matrix = linkage(X, 'ward')
figure = plt.figure(figsize=(7.5, 5))
dendrogram(
    linkage_matrix,
    color_threshold=0,
)

plt.tight_layout()
plt.show()
```
